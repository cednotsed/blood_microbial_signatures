# Calculating and visualising the proportion of human reads from Kraken2 reports

```r
setwd("/home/projects/14001280/PROJECTS/blood_microbiome/")
require(tidyverse)
require(ggplot2)
require(data.table)
```

```r
meta <- fread("data/SG10K_Health_metadata.n10714.16March2021.parsed.csv")
```

```r
unc_df <- tibble()

reports <- list.files("data/temp_files/06_reports/")
for (file in reports) {
#     file <- reports[15]
    file_path <- paste0("data/temp_files/06_reports/", file)
    
    if (file.size(file_path) != 0) {
        temp <- fread(paste0(file_path))
        filt <- temp %>% filter(V4 == "U")
        perc_unclassified <- filt$V1
    } else {
        perc_unclassified <- NA
    }
    
    npm_research_id <- unlist(strsplit(file, "\\."))[2]
    morsel <- tibble(perc_unclassified = perc_unclassified, npm_research_id = npm_research_id)
    unc_df <- unc_df %>%
    bind_rows(morsel)
}
```

```r
plt <- unc_df %>% 
    filter(!is.na(perc_unclassified)) %>% 
    left_join(meta, by = "npm_research_id") %>%
    ggplot(aes(x = source_cohort, y = perc_unclassified, fill = source_cohort)) +
    geom_boxplot(alpha = 0.5, outlier.shape = NA) +
    geom_point(color = "black", size = 1) +
    theme(axis.text.x = element_text(angle = 45, hjust = 1),
          axis.title.x = element_blank()) +
    labs(y = "% Unclassified Reads", fill = "Source Cohort")

plt
```

```r
ggsave("results/percentage_unclassified_reads.png", dpi = 600, plot = plt)
```
