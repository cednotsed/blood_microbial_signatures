### Load  data

```r
species_df <- fread("results/decontamination/read_matrix_n125.global_decontaminated.zeroed.csv")
species_RA <- fread("results/decontamination/RA_matrix_n125.global_decontaminated.zeroed.csv")
prev_df <- fread("results/decontamination/prevalence_RA0.01_read10.csv") %>%
    filter(npm_research_id %in% species_RA$npm_research_id)
meta <- fread("data/20210125_v3_release/20210203_all_traits_v3.txt")
n <- gsub("[^0-9A-Za-z///' ]", " ", colnames(meta))
n <- tolower(n)
n <- gsub(" ", "_", n)
colnames(meta) <- n
colnames(meta)

meta_filt <- meta %>%
    select(npm_research_id, age, cc_status, bmi, 
          height_cm, weight_kg, waistc_cm, hipc_cm, tc_mmol_l, hdl_mmol_l, tg_mmol_l,
          sbp, dbp, glu_fast_mmol_l, ins_fast_uiu_ml, hba1c, diab_med, htn_med, lipid_med) %>%
    filter(npm_research_id %in% species_df$npm_research_id) %>%
    drop_na()

head(meta_filt)
```

```r
meta_dat <- meta_filt %>%
    left_join(prev_df) 
summary(lm(data = meta_dat, formula = `Cutibacterium acnes` ~ bmi + height_cm + weight_kg + waistc_cm + hipc_cm))

summary(glm(data = meta_dat, formula = `Cutibacterium acnes` ~ tc_mmol_l + hdl_mmol_l + tg_mmol_l, family=binomial(link='logit')))
summary(glm(data = meta_dat, formula = `Cutibacterium acnes` ~ sbp + dbp, family=binomial(link='logit')))
summary(glm(data = meta_dat, formula = `Cutibacterium acnes` ~ diab_med + htn_med + lipid_med, family=binomial(link='logit')))
summary(glm(data = meta_dat, formula = `Cutibacterium acnes` ~ glu_fast_mmol_l + ins_fast_uiu_ml, family=binomial(link='logit')))
```
