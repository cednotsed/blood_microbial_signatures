### Load non-contaminant list

```r
# After thresholding
prev_df <- fread("results/decontamination/prevalence_RA0.005_read10.csv")
original_list <- colnames(prev_df)
original_list <- original_list[original_list != "npm_research_id"]
length(original_list)

# After max count filter
species_read_zeroed <- fread("results/decontamination/read_matrix_n117.global_decontaminated.zeroed.csv")
final_list <- colnames(species_read_zeroed)
final_list <- final_list[final_list != "npm_research_id"]
length(final_list)

human_df_parsed <- fread("results/decontamination/comparison/shaw_human_list_parsed.csv") %>%
    select(-Genus) %>%
    mutate(Human = ifelse(Human == "Yes", T, F))
poore_list <- fread("data/poore_et_al/poore_et_al_contaminant_list.csv")
blood_list <- read.csv("results/blood_culture_records/blood_culture_taxa_list.txt")$taxa

# Observed
after_df <- tibble(Species = final_list) %>%
    separate(Species, into = c("Genus", NA, NA), sep = " ", remove = F) %>%
    mutate(Genus = ifelse(grepl("virus", Species, ignore.case = T), NA, Genus)) %>%
    left_join(human_df_parsed)

after_df2 <- after_df %>%
    left_join(poore_list)

final_df <- after_df2 %>%
    mutate(Species = tolower(Species)) %>%
    mutate(blood = ifelse(Species %in% blood_list, T, F))

n_contam <- sum(final_df$category == "LIKELY CONTAMINANT", na.rm = T)
n_human <- sum(final_df$Human)
n_blood <- sum(final_df$blood)

obs_results <- tibble(prop_contam = n_contam / 117, prop_human = n_human / 117, prop_blood = n_blood / 117)

# Random sample
set.seed(66)

morsels <- foreach(i = seq(1000)) %do% {
    temp_list <- sample(original_list, 117)
    
    after_df <- tibble(Species = temp_list) %>%
        separate(Species, into = c("Genus", NA, NA), sep = " ", remove = F) %>%
        mutate(Genus = ifelse(grepl("virus", Species, ignore.case = T), NA, Genus)) %>%
        left_join(human_df_parsed)

    after_df2 <- after_df %>%
        left_join(poore_list)

    final_df <- after_df2 %>%
        mutate(Species = tolower(Species)) %>%
        mutate(blood = ifelse(Species %in% blood_list, T, F))

    n_contam <- sum(final_df$category == "LIKELY CONTAMINANT", na.rm = T)
    n_human <- sum(final_df$Human)
    n_blood <- sum(final_df$blood)

    tibble(prop_contam = n_contam / 117, prop_human = n_human / 117, prop_blood = n_blood / 117)
}

sampling_results <- bind_rows(morsels)
```

```r
sampling_results <- bind_rows(morsels)

obs_plot_df <- obs_results %>%
    pivot_longer(everything(), names_to = "classification", values_to = "prop")

plot_df <- sampling_results %>%
    pivot_longer(everything(), names_to = "classification", values_to = "prop") %>%
    mutate(classification = factor(classification, levels = c("prop_contam", "prop_blood", "prop_human")))

plot_df %>% fwrite("results/figure_source_data_nat_micro/obs_versus_random.csv")

plot_df %>%
    ggplot(aes(x = prop, fill = classification)) +
    facet_grid(rows = vars(factor(classification, levels = c("prop_contam", "prop_blood", "prop_human")))) +
    geom_histogram(bins = 100, color = "black") +
    scale_fill_manual(values = c("red", "coral2", "palegreen4")) +
    geom_vline(data = obs_plot_df, aes(xintercept = prop),
               color = "black", lty = "dashed") +
    theme_bw()

ggsave("results/decontamination/comparison/obs_versus_random.pdf", dpi = 600, height = 5, width = 8)
```

```r
col_names <- colnames(obs_results)
column <- col_names[1]

for (column in col_names) {
temp_res <- sampling_results %>%
    select(all_of(column)) %>%
    summarise(p.val = sum(get(column) >= obs_results[[column]]))

print(str_glue("{column}: {temp_res$p.val}"))
}
```

```r
sum(sampling_results$prop_contam <= obs_results$prop_contam)
sum(sampling_results$prop_human >= obs_results$prop_human)
sum(sampling_results$prop_blood >= obs_results$prop_blood)
```

```r
# Merge classifications with curated list
curated <- fread("results/decontamination/curated_n117_global_decontamination_stats.parsed.csv")

after_df <- tibble(taxa = final_list, Species = final_list) %>%
    separate(taxa, into = c("Genus", NA, NA), sep = " ", remove = F) %>%
    mutate(Genus = ifelse(grepl("virus", taxa, ignore.case = T), NA, Genus)) %>%
    left_join(human_df_parsed)

after_df2 <- after_df %>%
    left_join(poore_list)

final_df <- after_df2 %>%
    mutate(Species = tolower(taxa)) %>%
    mutate(blood = ifelse(Species %in% blood_list, T, F))

curated %>%
    left_join(final_df) %>%
    select(taxa, max_count, overall_prevalence, n_samples,
           org_group, site, microbiol_class,
           Human, category, blood) %>%
    fwrite("results/decontamination/curated_n117_global_decontamination_stats.parsed.with_classifications.csv")
# n_noncontam <- sum(is.na(final_df$category))
# n_human <- sum(final_df$Human)
# n_blood <- sum(final_df$blood)

# obs_results <- tibble(prop_noncontam = n_noncontam / 117, prop_human = n_human / 117, prop_blood = n_blood / 117)
```
