### Load  data

```r
prev_df <- fread("results/decontamination/prevalence_RA0.005_read10.csv")
nc <- fread("results/decontamination/curated_n117_global_decontamination_stats.parsed.csv")
nc <- unique(nc$taxa)
contams <- colnames(prev_df %>% select(-npm_research_id))
contams <- contams[!(contams %in% nc)]
length(contams)
```

```r
taxa_prev <- prev_df %>%
    pivot_longer(!npm_research_id, names_to = "taxa", values_to = "presence") %>%
    group_by(taxa) %>%
    summarise(prev = sum(presence) / n()) %>%
    arrange(desc(prev))

contam_subset <-taxa_prev %>% 
    filter(taxa %in% contams) %>%
    head(20)
taxa_prev %>% 
    filter(taxa %in% nc) %>%
    head(20)

fwrite(contam_subset %>% select(taxa), "results/irep_analysis/contaminants/contam_subset_list_n20.txt", col.names = F)
```

```r
species_df <- fread("results/decontamination/read_matrix.raw.zeroed.csv")
contam_reads <- species_df %>%
    pivot_longer(!npm_research_id, names_to = "taxa", values_to = "read_count") %>%
    filter(taxa %in% contams)

for (taxon in contam_subset$taxa) {
    temp_df <- contam_reads %>%
        filter(taxa == taxon) %>%
        arrange(desc(read_count)) %>%
        head(5)
    
    taxon_name <- gsub(" | sp. ", "_", taxon)
    
    fwrite(temp_df %>% select(npm_research_id), str_glue("data/irep_data/sample_lists/{taxon_name}_n5.txt"), col.names = F)
}
```

```r
species_df <- fread("results/decontamination/read_matrix_n125.global_decontaminated.zeroed.csv")
species_RA <- fread("results/decontamination/RA_matrix_n125.global_decontaminated.zeroed.csv")

replication_meta <- fread("results/irep_analysis/parsed_irep_results.csv") %>%
    filter(!is.na(bPTR)) %>%
    arrange(desc(bPTR))

global_stats <- fread("results/decontamination/curated_n125_global_decontamination_stats.csv")
top_20_prev_bact <- deframe(global_stats %>% 
                            filter(pathogen_type == "bacteria") %>%
                            arrange(desc(overall_prevalence)) %>%
                            head(20) %>% 
                            select(taxa))
top_5_prev_virus <- deframe(global_stats %>% 
                            filter(pathogen_type == "virus") %>%
                            arrange(desc(overall_prevalence)) %>%
                            head(5) %>% 
                            select(taxa))
top_5_prev_fungi <- deframe(global_stats %>% 
                            filter(pathogen_type == "fungi") %>%
                            arrange(desc(overall_prevalence)) %>%
                            head(5) %>% 
                            select(taxa))
top_20_max_bact <- deframe(global_stats %>% 
                           filter(pathogen_type == "bacteria") %>%
                           arrange(desc(max_count)) %>%
                           head(20) %>% 
                           select(taxa))
taxa_list <- unique(c(top_20_prev_bact, top_5_prev_virus, top_5_prev_fungi, top_20_max_bact))

length(taxa_list)
# replication_meta
# global_stats
```

#### No. of genera for wishlist species

```r
for (taxon in c(replication_meta$taxa, additional)) {
    samples_with_taxon <- species_df %>% 
        select(all_of(c("npm_research_id", taxon))) %>%
        filter(get(taxon) != 0) %>%
        arrange(desc(get(taxon)))
    samples_with_taxon <- samples_with_taxon$npm_research_id
    per_sample_RA <- species_RA %>% 
        filter(npm_research_id %in% samples_with_taxon) %>%
        pivot_longer(!npm_research_id, names_to = "taxa", values_to = "rel_a") %>%
        filter(rel_a != 0) %>%
        separate(taxa, into = c("genus", "species", "strain"), sep = " ", remove = F)

    plt <- per_sample_RA %>%
        group_by(npm_research_id) %>%
        summarise(n_genera = n()) %>%
        ggplot(aes(x = n_genera)) + 
            geom_histogram() +
        labs(title = taxon)
    
    print(plt)
           
}
```
