# Calculating and visualising the proportion of XXX reads from Kraken2 reports

```r
setwd("/home/projects/14001280/PROJECTS/blood_microbiome/")
require(tidyverse)
require(ggplot2)
require(data.table)
```

```r
meta <- fread("data/SG10K_Health_metadata.n10714.16March2021.parsed.csv")
reports <- list.files("data/temp_files/06_reports/")
```

```r
get_df <- function(x) {
    df <- tibble()

    for (file in reports) {
#         file <- reports[15]
#         x <- "unclassified"
        file_path <- paste0("data/temp_files/06_reports/", file)

        if (file.size(file_path) != 0) {
            temp <- fread(paste0(file_path))
            filt <- temp %>% filter(V6 == x)
            perc_x <- filt$V1
        } else {
            perc_x <- NA
        }

        npm_research_id <- unlist(strsplit(file, "\\."))[2]
        morsel <- tibble(perc_x = perc_x, npm_research_id = npm_research_id)
        df <- df %>%
        bind_rows(morsel)
    }
    
    return(df)
}
```

```r
visualise_perc <- function(x) {
    df <- get_df(x)
    plt <- df %>% 
        filter(!is.na(perc_x)) %>% 
        left_join(meta, by = "npm_research_id") %>%
        ggplot(aes(x = source_cohort, y = perc_x, fill = source_cohort)) +
        geom_boxplot(alpha = 0.5, outlier.shape = NA) +
        geom_point(color = "black", size = 1) +
        theme(axis.text.x = element_text(angle = 45, hjust = 1),
              axis.title.x = element_blank()) +
        labs(y = paste("% Reads", x), fill = "Source Cohort")

    return(list(plt = plt, df = df))
}

res_unc <- visualise_perc("unclassified")
res_human <- visualise_perc("Homo sapiens")
```
